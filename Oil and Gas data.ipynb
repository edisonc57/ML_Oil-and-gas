Tabla de estadísticas descriptivas 

*count: Es el número de datos o muestras disponibles para cada variable, en este caso, 1100.

*mean: Es el valor promedio de cada variable.
Ejemplo: El valor promedio de CGR es 22.06.

*std: Es la desviación estándar, que mide la dispersión de los valores respecto al promedio.
Ejemplo: La desviación estándar de NPHI es 0.054.

*min: Es el valor mínimo registrado para cada variable.
Ejemplo: El valor mínimo de RHOB es -9999.0, que podría indicar valores faltantes o erróneos.

*25% (Primer cuartil): Es el valor debajo del cual se encuentra el 25% de los datos.
Ejemplo: El 25% de los datos de DT están por debajo de 80.91.

*50% (Mediana): Es el valor medio o mediana, donde el 50% de los datos están por encima y el 50% por debajo.
Ejemplo: El valor mediano de CGR es 19.48.

*75% (Tercer cuartil): Es el valor debajo del cual se encuentra el 75% de los datos.
Ejemplo: El 75% de los datos de RHOB están por debajo de 2.3947.

*max: Es el valor máximo registrado para cada variable.
Ejemplo: El valor máximo de DT es 177.69.

Evaluando la data perdida o nula

1. isnull()
Propósito: Detectar valores nulos (NaN).

2. notnull()
Propósito: Detectar valores no nulos.

3. dropna()
Propósito: Eliminar las filas o columnas que contienen valores nulos (NaN).

4. fillna()
Propósito: Rellenar los valores nulos (NaN) con algún valor específico.

La función msno.matrix(df) pertenece a la biblioteca missingno (abreviada como msno), que es una herramienta visual para analizar valores faltantes o nulos en los datos de un DataFrame de pandas.

```python
pip install missingno
```

Barra blanca: Representa los valores faltantes (NaN) en las filas del DataFrame.

Barra gris: Representa los valores presentes (datos no nulos) en el DataFrame.

```python
import missingno as msno

msno.matrix(df,figsize=(10,5), fontsize = 10, color= (0.6, 0.6, 0.5))
```

```python
msno.bar(df,figsize= (10,5), fontsize= 10, color = (0.5,0.7,0.5),  sort= 'ascending')
```
